```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, TimeDistributed, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from gensim.models import KeyedVectors
```

```python
data = [
    ("Saya mau memesan 10 gelas motif Kotak-kotak di Jakarta Selatan", 
     [("10", "B-Jumlah"), ("gelas", "B-Nama"), ("motif", "O"), 
      ("Kotak-kotak", "B-Motif"), ("di", "O"), ("Jakarta Selatan", "B-Lokasi")]),
    ("Saya membeli 5 piring dengan motif bunga di Bandung",
     [("5", "B-Jumlah"), ("piring", "B-Nama"), ("motif", "O"),
      ("bunga", "B-Motif"), ("di", "O"), ("Bandung", "B-Lokasi")]),
    ("Saya mau memesan 5 piring dan 4 gelas dengan motif floral",
     [("5", "B-Jumlah"), ("4", "B-Jumlah"), ("piring", "B-Nama"), ("gelas", "B-Nama"),
      ("floral", "B-Motif"), ("dengan", "O"),]),
    ("Aku membutuhkan 3 kursi bermotif lingkaran untuk dikirim ke Surabaya",
     [("3", "B-Jumlah"), ("kursi", "B-Nama"), ("bermotif", "O"),
      ("lingkaran", "B-Motif"), ("untuk", "O"), ("dikirim", "O"),
      ("ke", "O"), ("Surabaya", "B-Lokasi")])
]
```

```python
# Tokenisasi dan pelabelan
def prepare_data(data):
    sentences = []
    labels = []
    for text, entities in data:
        tokens = text.split()
        tag_seq = ["O"] * len(tokens)
        for entity, label in entities:
            if entity in tokens:
                idx = tokens.index(entity)
                tag_seq[idx] = label
        sentences.append(tokens)
        labels.append(tag_seq)
    return sentences, labels

# Preprocessing data
sentences, labels = prepare_data(data)

# Mapping kata ke indeks
word2idx = {w: i + 1 for i, w in enumerate(set(word for sent in sentences for word in sent))}
label2idx = {l: i for i, l in enumerate(set(label for lbl_seq in labels for label in lbl_seq))}
idx2label = {i: l for l, i in label2idx.items()}

# Data encoding
X = [[word2idx[word] for word in sent] for sent in sentences]
y = [[label2idx[label] for label in lbl_seq] for lbl_seq in labels]

# Hitung panjang maksimum sekuens
max_len = max(len(x) for x in X)

# Padding
X = pad_sequences(X, maxlen=max_len, padding='post')
y = pad_sequences(y, maxlen=max_len, padding='post')
y = np.array([np.eye(len(label2idx))[lbl_seq] for lbl_seq in y])  # One-hot encoding labels

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from gensim.scripts.glove2word2vec import glove2word2vec

# Path ke file GloVe asli pre-trained embeddings menggunakan glove
glove_input_file = "glove.6B.100d.txt"
word2vec_output_file = "glove.6B.100d.word2vec.txt"

# Konversi
glove2word2vec(glove_input_file, word2vec_output_file)
```

```python

# Load pre-trained embeddings (GloVe)
embedding_path = "glove.6B.100d.word2vec.txt"  # File hasil konversi
embedding_dim = 100
embedding_model = KeyedVectors.load_word2vec_format(embedding_path, binary=False)

# Buat embedding matrix
embedding_matrix = np.zeros((len(word2idx) + 1, embedding_dim))
for word, i in word2idx.items():
    if word in embedding_model:
        embedding_matrix[i] = embedding_model[word]
    else:
        embedding_matrix[i] = np.random.normal(scale=0.6, size=(embedding_dim,))  # Fallback untuk kata tidak ditemukan

# Definisikan embedding layer
embedding_layer = Embedding(
    input_dim=len(word2idx) + 1,
    output_dim=embedding_dim,
    weights=[embedding_matrix],
    input_length=max_len,
    trainable=False  # Embedding tidak dilatih ulang
)

# Model
model = Sequential([
    embedding_layer,
    Bidirectional(LSTM(units=64, return_sequences=True, dropout=0.3, recurrent_dropout=0.3)),
    TimeDistributed(Dense(len(label2idx), activation='softmax'))
])

# Compile dan latih model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
history = model.fit(X_train, y_train, batch_size=16, epochs=30, validation_split=0.1)
```

```python
history = model.fit(X_train, y_train, batch_size=16, epochs=30, validation_split=0.1)
```

```python
# Prediksi
y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=-1)
y_true = np.argmax(y_test, axis=-1)

# Decode label
y_pred_labels = [[idx2label[idx] for idx in pred_seq] for pred_seq in y_pred]
y_true_labels = [[idx2label[idx] for idx in true_seq] for true_seq in y_true]

# Flatten labels
y_pred_flat = [lbl for seq in y_pred_labels for lbl in seq]
y_true_flat = [lbl for seq in y_true_labels for lbl in seq]

# Evaluasi
print(classification_report(y_true_flat, y_pred_flat, labels=list(label2idx.keys())))
```

```python
# Testing dengan input baru
test_text = "Saya mau memesan 5 piring dan 4 gelas dengan motif floral"
test_tokens = test_text.split()
test_indices = [word2idx.get(word, 0) for word in test_tokens]
test_indices = pad_sequences([test_indices], maxlen=max_len, padding='post')

# Prediksi
predicted = model.predict(test_indices)
predicted = np.argmax(predicted, axis=-1)

# Decode hasil prediksi
predicted_labels = [idx2label[idx] for idx in predicted[0][:len(test_tokens)]]

# Tampilkan hasil
print("\nInput Text:")
print(test_text)
print("\nTokenized Words and Predicted Labels:")
for token, label in zip(test_tokens, predicted_labels):
    print(f"{token}: {label}")
```
